```python
%%sql

/* 1. Create the table with STRICT types */
CREATE TABLE school_supplies (
    supply_id INTEGER PRIMARY KEY,
    item_name TEXT NOT NULL,
    category TEXT,
    price REAL NOT NULL
) STRICT;
```

```python
%%sql

/* 2. Insert at least 3 items */
INSERT INTO school_supplies (item_name, category, price)
VALUES
('Notebook', 'Paper Goods', 4.99),
('Mechanical Pencil Pack', 'Writing Tools', 6.50),
('Backpack', 'Storage', 39.99),
('Calculator', 'Electronics', 89.99);
```

```python
%%sql

/* Show all supplies that cost more than $10 */
SELECT *
FROM school_supplies
WHERE price > 10;
```

```python
%%sql

/* Update price of backpack */
UPDATE school_supplies
SET price = 34.99
WHERE item_name = 'Backpack';
```

```python
%%sql

/* Delete one item */
DELETE FROM school_supplies
WHERE supply_id = 1;
```
