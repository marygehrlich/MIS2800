## YOUR ANSWERS TO PART 1 BELOW

Question 1: Moving to Amazon Redshift allowed Lyft to analyze massive amounts of historical ride data instead of just storing it. This helped them identify ride patterns, demand trends, and routing efficiencies, which enabled them to create new products like Lyft Line. Instead of just tracking rides, they could use data strategically to drive innovation and revenue growth.


Question 2:Lyft uses different systems because operational databases (OLTP) handle fast, real-time ride transactions, while analytical systems (OLAP) handle large-scale data analysis. Tools like Amazon DynamoDB support quick app transactions, while Redshift is optimized for complex queries and long-term planning. Using one database for both would slow down the app and reduce performance.

## YOUR ANSWERS TO PART 2 BELOW

Question 1:Columnar storage is more important for analysts because it makes large data queries much faster. In Amazon Redshift, data is stored by column, so when an analyst only needs a few columns, the system reads less data and runs faster.Normalization matters more for transaction systems, but in data warehouses, speed and performance for big queries are more important.

Question 2: In Amazon Redshift, the Leader Node sends parts of a query to multiple Compute Nodes. Since the compute nodes process data at the same time, BI reports run much faster than if one machine handled everything alone.
